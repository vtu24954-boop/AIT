```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Employee.csv")
print(df)
```

```python
df['PaymentTier'].value_counts().plot(kind='bar')
plt.title("Bar Chart of PaymentTier")
plt.xlabel("PaymentTier")
plt.ylabel("Count")
plt.show()
```

```python
df['PaymentTier'].value_counts().plot(kind='pie')
plt.title("pie Chart of PaymentTier")
plt.xlabel("PaymentTier")
plt.ylabel("Count")
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("pie Chart of PaymentTier")
plt.xlabel("PaymentTier")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df.index,df['Age'])
plt.title("Scatter Plot of Age")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
plt.plot(df['JoiningYear'])
plt.title("Line Plot of Joining Year")
plt.xlabel("Index")
plt.ylabel("Year")
plt.show()
```

```python
plt.plot(df['Age'],'|')
plt.title("Strip Plot of Age")
plt.xlabel("Age")
plt.show()
```

```python

plt.scatter(df['ExperienceInCurrentDomain'],[1]*len(df))
plt.title("Swaram-like plot of Experince")
plt.xlabel("Experince in Current Domain")
```
